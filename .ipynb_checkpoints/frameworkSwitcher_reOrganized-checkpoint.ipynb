```python
import os
os.chdir('/lustre/ARRAKIHS/wissarut/')

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob
from astropy.io import fits
from astropy.constants import c
path_filters='jdap_arrakihs/total_transimission_18122023/'
```

```python
path_filters='jdap_arrakihs/total_transimission_18122023/'
def get_filters_dict():
    filters_names=glob.glob(path_filters+'*')
    filter_VIS1=pd.read_csv(filters_names[0],header=0,names=['Wavelength(AA)','Transmission'])
    filter_VIS2=pd.read_csv(filters_names[1],header=0,names=['Wavelength(AA)','Transmission'])
    filter_NIR2=pd.read_csv(filters_names[2],header=0,names=['Wavelength(AA)','Transmission'])
    filter_NIR1=pd.read_csv(filters_names[3],header=0,names=['Wavelength(AA)','Transmission'])

    filter_NIR1NIR2=filter_NIR1.copy(deep=True)
    filter_VIS2VIS1=filter_VIS2.copy(deep=True)
    filter_NIR1NIR2['Transmission']=filter_NIR1['Transmission']-filter_NIR2['Transmission']
    filter_VIS2VIS1['Transmission']=filter_VIS2['Transmission']-filter_VIS1['Transmission']
    ### data as dict ###
    filters_dict={'VIS1':filter_VIS1,'VIS2-VIS1':filter_VIS2VIS1,'NIR1-NIR2':filter_NIR1NIR2,'NIR2':filter_NIR2}
    
    return filters_dict

filters_dict = get_filters_dict()


def plot_filters_dict(filters_dict,plot_name):

    for name,data in filters_dict.items():
        plt.plot(data['Wavelength(AA)'],data['Transmission'],label=name,alpha=0.5,linestyle='--')
        plt.fill_between(data['Wavelength(AA)'],data['Transmission'],alpha=0.2)
    plt.title(plot_name)
    plt.xlabel('Wavelength (AA)')
    plt.ylabel('Transmission')
    plt.legend()
    plt.show()
    

plot_filters_dict(filters_dict,'ARRAKIHS Filters')
```

```python
import os
import glob
import pandas as pd
from itertools import combinations

path_filters = 'jdap_arrakihs/total_transimission_18122023/'

def get_filters_dict(path=path_filters):
    filters_dict = {}
    
    #Get filters name
    filter_files = sorted(glob.glob(os.path.join(path, '*')))
    
    # Optional: extract cleaner names from filenames
    for file in filter_files:
        # Get just the base name, without extension
        name = os.path.splitext(os.path.basename(file))[0]
        df = pd.read_csv(file, header=0, names=['Wavelength(AA)', 'Transmission'])
        filters_dict[name] = df

    return filters_dict

# Usage
filters_dict = get_filters_dict()
```

```python
plot_filters_dict(filters_dict,'ARRAKIHS Filters')
```

```python

def filters_subtract(filters_dict, subtract_pairs):
    """
    subtract_pairs: list of tuples like [('VIS2', 'VIS1'), ('NIR1', 'NIR2')]
    """
    filters_diff = {}

    for name1, name2 in subtract_pairs:
        if name1 in filters_dict and name2 in filters_dict:
            df1 = filters_dict[name1]
            df2 = filters_dict[name2]

            # Check if wavelengths match
            if not df1['Wavelength(AA)'].equals(df2['Wavelength(AA)']):
                print(f"Skipping {name1}-{name2}: Wavelength grids do not match.")
                continue

            diff_df = df1.copy(deep=True)
            diff_df['Transmission'] = df1['Transmission'] - df2['Transmission']
            diff_name = f"{name1}-{name2}"
            filters_diff[diff_name] = diff_df
        else:
            print(f"Skipping {name1}-{name2}: One or both filters not found.")

    # Return a new combined dictionary
    return {**filters_dict, **filters_diff}

# Example usage
filters_dict = get_filters_dict()

# Define pairs to subtract (customizable!)
subtract_pairs = [('VIS2', 'VIS1'), ('NIR1', 'NIR2')]

# Get updated dictionary including difference filters
filters_dict = filters_subtract(filters_dict, subtract_pairs)
```

```python
plot_filters_dict(filters_dict,'ARRAKIHS Filters')
```
